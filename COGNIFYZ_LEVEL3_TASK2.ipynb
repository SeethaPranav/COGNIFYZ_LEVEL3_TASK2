## Identify the Restaurants with the Highest and Lowest Number of Votes

```python
# Identify the restaurant(s) with the highest and lowest number of votes
max_votes = data['Votes'].max()
min_votes = data['Votes'].min()

# Restaurants with the highest number of votes
max_votes_restaurants = data[data['Votes'] == max_votes]

# Restaurants with the lowest number of votes
min_votes_restaurants = data[data['Votes'] == min_votes]

# Display results
print("Restaurant(s) with the highest number of votes:")
print(max_votes_restaurants[['Restaurant Name', 'Votes']])

print("\nRestaurant(s) with the lowest number of votes:")
print(min_votes_restaurants[['Restaurant Name', 'Votes']])
```

## Analyze if there is a Correlation Between the Number of Votes and the Rating of a Restaurant

### 1. Plot a scatter plot to visualize the relationship.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['Votes'], y=data['Aggregate rating'], alpha=0.6)

plt.title('Relationship Between Number of Votes and Aggregate Rating')
plt.xlabel('Number of Votes')
plt.ylabel('Aggregate Rating')
plt.grid(True)
plt.show()
```

### 2. Calculate the correlation between the two variables.

```python
# Calculate the correlation between the number of votes and aggregate rating
correlation_votes_rating = data[['Votes', 'Aggregate rating']].corr().iloc[0, 1]
print(f"Correlation between Number of Votes and Aggregate Rating: {correlation_votes_rating:.2f}")
```

## RESULTS

#### The restaurant with the highest number of votes is Toit, with 10,934 votes, while several restaurants have zero votes, indicating either no reviews or very limited feedback.

#### The correlation between the number of votes and aggregate rating is 0.31, suggesting a moderate positive relationship between the two, where restaurants with more votes tend to have slightly higher ratings, but the relationship is not very strong.
